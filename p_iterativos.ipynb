# Ejercicio 1

```python
%autoreload 2
from src import gauss_jacobi, gauss_seidel
A = [[2,1],[3,2]]
b = [5,8]
n = len(b)

x, trayectoria = gauss_seidel(A=A, b=b, x=[0]*n, tol = 1e-5, max_iter=20)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    return 5-2*x

def f2(x):
    return (8-3*x)/2

puntos = np.array(trayectoria).squeeze()

x = np.linspace(-2,5,100)
x1 = puntos[:,0]
x2 = puntos[:,1]

# Dibujar trajectoria en el espacio 2D
plt.plot(x, f1(x), "blue")
plt.plot(x, f2(x), "green")
plt.plot(x1, x2, "o:", markersize=4)
plt.plot(x1[0], x2[0], "gd", markersize=8, label="$X_0$", alpha=0.8)
plt.plot(x1[-1], x2[-1], "ro", markersize=8, label="$X_{sol}$", alpha=0.8)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trayectoria con Gauss Seidel")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
